# Xpath

```
//: 최상위 엘리먼트
* : 자손 태그 검색 => div form #띄어 쓰기 = 자손태그 검색 꺽쇠는 자식태그
/ : 자식 태그 검색 => div > form
td[2]: td 태그 중에서 2번째 태그를 선택 => td:nth-child(2)
```

````python
#xpath

xpath = '''//*[@id="___gcse_0"]/div/form/table/tbody/tr/td[2]/button'''
some_tag = driver.find_element_by_xpath(xpath)
some_tag.click()
````

````python
css = "#___gcse_0 > div > form > table > tbody > tr > td:nth-child(2) > button"
some_tag = driver.find_element_by_css_selector(css)
some_tag.click()
````
